```python
import matplotlib.pyplot as plt

df['age'].value_counts().sort_index().plot(kind='bar')

plt.title("Age Distribution of Players")
plt.xlabel("Age")
plt.ylabel("Number of Players")

plt.show()
```

```python
df.groupby('team')['tournament_rating'].mean().sort_values(ascending=False)
```

```python
df.groupby('team')['tournament_rating'].mean().sort_values(ascending=False).plot(
    kind='bar',
    figsize=(25,5),
    color='red'
)

plt.title("Average Tournament Rating by Team")
plt.xlabel("Team")
plt.ylabel("Average Tournament Rating")
plt.xticks(rotation=45)

plt.show()
```

# Insight: The chart compares the average tournament rating of each team. Teams with higher average ratings performed better overall in the tournament, while teams with lower average ratings showed relatively weaker performance.

```python
top_players = df.nlargest(10, 'tournament_rating')

top_players[['player_name', 'team', 'tournament_rating']]
```

```python
top_players.plot(
    x='player_name',
    y='tournament_rating',
    kind='bar',
    figsize=(10,5)
)

plt.title("Top 10 Highest Rated Players")
plt.xlabel("Player")
plt.ylabel("Tournament Rating")
plt.xticks(rotation=45)

plt.show()
```

```python
top_goals = df.nlargest(10, 'total_goals_tournament')

top_goals[['player_name', 'team', 'total_goals_tournament']]
```

```python
import matplotlib.pyplot as plt

top_goals.plot(
    x='player_name',
    y='total_goals_tournament',
    kind='barh',
    figsize=(12,6),
    color='teal',
    edgecolor='black'
)

plt.title("Top 10 Goal Scorers")
plt.xlabel("Goals")
plt.ylabel("Player")
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.show()
```

```python
top_assists = df.nlargest(10, 'total_assists_tournament')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(
    top_assists['player_name'],
    top_assists['total_assists_tournament'],
    s=150
)

plt.title("Top 10 Assist Providers")
plt.xlabel("Player")
plt.ylabel("Total Assists")
plt.xticks(rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(38,25))   

sns.heatmap(
    df.select_dtypes(include='number').corr(),
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)

plt.title("Correlation Heatmap", fontsize=18, fontweight='bold')

plt.show()
```

```python
top_pom= df.nlargest(10, 'player_of_match_awards')
```

```python
top_pom.plot(
    x='player_name',
    y='player_of_match_awards',
    kind='barh',
    figsize=(12,6),
    color='gold',
    edgecolor='black'
)

plt.title("Top 10 Players by Player of the Match Awards")
plt.xlabel("Awards")
plt.ylabel("Player")
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.show()
```

```python
df.groupby('position')['total_goals_tournament'].mean().sort_values(ascending=False).plot(
    kind='bar',
    figsize=(15,9),
    color='crimson',
    edgecolor='black'
)

plt.title("Average Goals by Position")
plt.xlabel("Position")
plt.ylabel("Average Goals")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
